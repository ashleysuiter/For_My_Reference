## Handling Missing Values

Pandas represents missing values as `None` or `np.NaN`.

It is important to distinguish between several classes of missing data and handle it accordingly:

- Data that was never collected
- Data that was handled incorrectly at some point (possibly by you!)
- A "null" value, e.g. "zero", "false", "not applicable", or "entered an empty string"

Truly missing data can be handled in three ways:

- Delete the row
- Delete the column
- Fill in the missing values, e.g. with a mean or median or a value inferred from the other columns

If missingness is correlated with the value of the variable in question (e.g. patients who aren't getting better on a trial drug tend not to come back after the first dose, so their results aren't recorded), then you are in trouble.

```python
# missing values are usually excluded in calculations by default
drinks.loc[:,'continent'].value_counts()
```

```python
# includes missing values
drinks.loc[:,'continent'].isna()
```

```python
# find missing values in a Series
# True if missing, False if not missing
```

```python
# includes missing values
drinks.loc[drinks.loc[:,'continent'].isna(),:]
```

```python
# count the missing values - sum() works because True is 1 and False is 0
drinks.loc[:,'continent'].isna().sum()
```

```python
# True if not missing, False if missing
drinks.loc[:,'continent'].notna()
```

```python
# only show rows where continent is not missing
drinks.loc[drinks.loc[:,'continent'].notna(),:]
```

## Applying Functions

The `apply` method calls the provided function on each item in a Pandas Series or along the specified dimension of a Pandas DataFrame.

```python
# Series example
def is_under_30(num):
    return num<30
users.loc[:,'age'].apply(is_under_30)
```

You can use the `lambda` keyword to create an "anonymous function" on the fly.

(`lambda` functions are a general Python construction, not unique to the `apply` context, but they are common in that context.)

```python
# Same thing, but use `lambda` to define the function inside the call to `apply`
users.loc[:,'age'].apply(lambda age:age<30)
```

```python
# Applying across the columns of a DataFrame to identify whether each user is a male under 30.
```

It is a bit confusing that to get the result of applying a function `my_func` to each row, you need to run `df.apply(my_func, axis='columns')` rather than `df.apply(my_func, axis='index')`. But the `axis` argument specifies that you want to apply the function in question **along** the specified dimension: 

- `axis="columns"` means to apply "**along** the columns axis," so the ouput has one result per row.
- `axis="index"` means "apply "**along** the index (rows) axis," so the ouput has one result per column.
