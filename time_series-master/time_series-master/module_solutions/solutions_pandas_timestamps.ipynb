# Pandas Timestamps

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head()
```

```python
# Convert times from str to pandas Timestamp objects
# /scrub/
ufo.loc[:, 'Time'] = pd.to_datetime(ufo.loc[:, 'Time'])
ufo.head()
```

```python
# Set times as index
# /scrub/
ufo.set_index('Time', inplace=True)
ufo.head()
```

```python
# Get all data from 1936
# /scrub/
ufo.loc['1936', :]
```
